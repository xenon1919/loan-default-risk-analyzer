# Import necessary packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Load the dataset

```python
df = pd.read_csv("../data/raw/loan_default_dataset_v2.csv")
```

## Class Probability (Base Rate Reality)

```python
df["default"].value_counts(normalize=True)
```

## Feature Distributions (One by One)

```python
plt.hist(df["annual_income"], bins=30)
plt.title("Annual Income Distribution")
plt.show()
```

## Conditional Probability

### Probability of default given low income

```python
low_income = df[df["annual_income"] < df["annual_income"].median()]
low_income["default"].mean()
```

### Probability of default given high interest rate

```python
high_interest = df[df["interest_rate"] > df["interest_rate"].median()]
high_interest["default"].mean()
```
