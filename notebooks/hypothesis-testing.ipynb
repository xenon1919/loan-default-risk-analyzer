# Import necessary packages

```python
import pandas as pd
import numpy as np
from scipy import stats
```

# Load the dataset

```python
df = pd.read_csv("../data/raw/loan_default_dataset_v2.csv")
```

```python
default_0 = df[df["default"] == 0]
default_1 = df[df["default"] == 1]
```

# Test 1: Income Difference (t-test)

### H₀ (Null): Mean income of defaulters = non-defaulters

### H₁ (Alt): Mean income of defaulters ≠ non-defaulters

```python
stats.ttest_ind(
    default_0["annual_income"],
    default_1["annual_income"],
    equal_var=False
)
```

# Test 2: Interest Rate Difference (t-test)

```python
stats.ttest_ind(
    default_0["interest_rate"],
    default_1["interest_rate"],
    equal_var=False
)
```

# Test 3: Credit Score Difference (t-test)

```python
stats.ttest_ind(
    default_0["credit_score"],
    default_1["credit_score"],
    equal_var=False
)
```

# Test 4: Employment Years

```python
stats.ttest_ind(
    default_0["employment_years"],
    default_1["employment_years"],
    equal_var=False
)
```
